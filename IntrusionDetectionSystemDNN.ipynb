# Training the Algorithm

```python
# Algorithmic training
np.random.seed(41)

def train (features_train, targets_train, features_test, targets_test):
    # Hyperparameters
    n_hidden = 1  # number of hidden units
    epochs = 5
    learnrate = 0.0001

    n_records, n_features = features_train.shape
    bias = np.full((n_features, n_hidden), 0.9)
    last_loss = None
    # Initialize weights
    # First Layer
    weights_input_hidden1 = np.random.normal(scale=1 / n_features ** .5,
                                            size=(n_features, n_hidden))
    
    # Second Layer
    weights_hidden1_hidden2 = np.random.normal(scale=1 / n_features ** .5,
                                            size=(n_hidden, n_hidden))
    
    # Third Layer
    weights_hidden2_hidden3 = np.random.normal(scale=1 / n_features ** .5,
                                            size=(n_hidden, n_hidden))
    # Output Layer
    weights_hidden3_output = np.random.normal(scale=1 / n_features ** .5,
                                             size=(n_hidden,1))
    
    for e in range(epochs):
        del_w_input_hidden1 = np.zeros(weights_input_hidden1.shape)
        del_w_hidden1_hidden2 = np.zeros(weights_hidden1_hidden2.shape)
        del_w_hidden2_hidden3 = np.zeros(weights_hidden2_hidden3.shape)
        del_w_hidden3_output = np.zeros(weights_hidden3_output.shape)
        
        counter = 0
        for x, y in zip(features_train, targets_train):
            ## Forward pass ##
            # Calculate the output
            hidden1_input = np.dot(x, weights_input_hidden1) + bias
            hidden1_output = sigmoid(hidden1_input)
            
            hidden2_input = np.dot(hidden1_output, weights_hidden1_hidden2)
            hidden2_output = sigmoid(hidden2_input) + bias
            
            hidden3_input = np.dot(hidden2_output, weights_hidden2_hidden3) + bias
            hidden3_output = sigmoid(hidden3_input)
            
            
            output = sigmoid(np.dot(hidden3_output,
                                    weights_hidden3_output) + bias)

            ## Backward pass ##
            #Calculate the network's prediction error
            error = y - output

            # Calculate error term for the output unit
            output_error_term = error * output * (1 - output)

            ## propagate errors backwardly from the output to the input layer
        
            # Calculate the third hidden layer's contribution to the error
            hidden3_error = np.dot(output_error_term, weights_hidden3_output)
            
            # Calculate the error term for the third hidden layer
            hidden3_error_term = hidden3_error * hidden3_output * (1 - hidden3_output)
            
            # Calculate the second hidden layer's contribution to the error
            hidden2_error = np.dot(hidden3_error_term, weights_hidden2_hidden3)
            
            # Calculate the error term for the second layer
            hidden2_error_term = hidden2_error * hidden2_output * (1 - hidden2_output)
            
            # Calculate the first hidden layer's contribution to the error
            hidden1_error = np.dot(hidden2_error_term, weights_hidden1_hidden2)
            
            # Calculate the error term for the first layer
            hidden1_error_term = hidden1_error * hidden1_output * (1 - hidden1_output)
            
            # Calculate the change in weight(deltaW)
            del_w_hidden3_output += output_error_term * hidden3_output
            del_w_hidden2_hidden3 += hidden3_error_term *  hidden2_output
            del_w_hidden1_hidden2 += hidden2_error_term * hidden1_output
            del_w_input_hidden1 += hidden1_error_term * x[:, None]
            
            # Calculate the change in bias(deltaB)


        # Update weights
        weights_input_hidden1 += learnrate * del_w_input_hidden1 / n_records
        weights_hidden1_hidden2 += learnrate * del_w_hidden1_hidden2 / n_records
        weights_hidden2_hidden3 += learnrate * del_w_hidden2_hidden3 / n_records
        weights_hidden3_output += learnrate * del_w_hidden3_output / n_records

        # Printing out the mean square error (loss) on the training set
        if e % (epochs / 10) == 0:
            hidden1_input = np.dot(x, weights_input_hidden1) + bias
            hidden1_output = sigmoid(hidden1_input)
            
            hidden2_input = np.dot(hidden1_output, weights_hidden1_hidden2) + bias
            hidden2_output = sigmoid(hidden2_input)
            
            hidden3_input = np.dot(hidden2_output, weights_hidden2_hidden3) + bias
            hidden3_output = sigmoid(hidden3_input)
            
            
            output = sigmoid(np.dot(hidden3_output,
                                    weights_hidden3_output) + bias)
            
            # get the loss
            loss = np.mean((output - targets_train) ** 2)

            # compare the last loss and the current one
            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            else:
                print("Train loss: ", loss)
            last_loss = loss

        ## if validation data set is extracted along with the training and test sets
        #if  e % (epochs / 50) == 0:
        #    hidden = sigmoid(np.dot(features_validation, weights_input_hidden))
        #    out = sigmoid(np.dot(hidden, weights_hidden_output))
        #    loss = np.mean((out - targets))

        #    if last_val_loss and last_val_loss < loss:
        #        print("Validation Loss: ", loss, " WARNING - Loss Increasing")

    # Calculate accuracy on test data
    #hidden = sigmoid(np.dot(features_test, weights_input_hidden))
    #out = sigmoid(np.dot(hidden, weights_hidden_output))
    hidden1_input = np.dot(features_test, weights_input_hidden1) + bias
    hidden1_output = sigmoid(hidden1_input)
            
    hidden2_input = np.dot(hidden1_output, weights_hidden1_hidden2) + bias
    hidden2_output = sigmoid(hidden2_input)
            
    hidden3_input = np.dot(hidden2_output, weights_hidden2_hidden3) + bias
    hidden3_output = sigmoid(hidden3_input)
            
            
    output = sigmoid(np.dot(hidden3_output,
                                    weights_hidden3_output) + bias)
    predictions = output > 0.8
    accuracy = np.mean(predictions == targets_test)
    print("Prediction accuracy: {:.3f}".format(accuracy))
    print(predictions)
    print(output)
```
